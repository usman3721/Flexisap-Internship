```python
# Exercise 2

def sum_recursive(n):
  """Calculates the sum of numbers from 0 to n recursively.

  Args:
      n (int): The upper limit of the summation (inclusive).

  Returns:
      int: The sum of numbers from 0 to n.
  """
  if n == 0:
    return 0
  else:
    return n + sum_recursive(n-1)

# Calculate the sum from 0 to 15
result = sum_recursive(15)
print(result)


```

```python
# Exercise 3A

# Assuming you have a CSV file containing the data
data = pd.read_csv('company_sales_data.csv')

# Display the first few rows of the data to understand its structure
print(data.head())

# Exercise 1: Read Total profit of all months and show it using a line plot
plt.figure(figsize=(10, 5))
plt.plot(data['month_number'], data['total_profit'], marker='o', linestyle='-', color='b')
plt.title('Total Profit per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

```python
# Exercise 3B
# Exercise 2: Read Bathing soap and facewash sales of all months and display them using Subplots
plt.figure(figsize=(12, 8))

# Subplot 1: Bathing Soap Sales
plt.subplot(2, 1, 1)
plt.plot(data['month_number'], data['bathingsoap'], marker='o', linestyle='-', color='g')
plt.title('Bathing Soap Sales per Month')
plt.xlabel('Month Number')
plt.ylabel('Bathing Soap Sales')
plt.grid(True)

# Subplot 2: Facewash Sales
plt.subplot(2, 1, 2)
plt.plot(data['month_number'], data['facewash'], marker='o', linestyle='-', color='r')
plt.title('Facewash Sales per Month')
plt.xlabel('Month Number')
plt.ylabel('Facewash Sales')
plt.grid(True)

plt.tight_layout()
plt.show()
```
